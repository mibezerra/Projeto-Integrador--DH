```python
def dados_ausentes(rotten): # uma função de dados ausentes para verificar se existe alum dado faltante
    total = rotten.isnull().sum() # aqui ele somou todos os lugares que não tinham valores 
    percentual = (rotten.isnull().sum()/rotten.isnull().count()*100) # calculou o percentual desses valores
    total_percentual = pd.concat([total, percentual], axis=1, keys=['total', 'percentual']) 
    types = []
    for col in rotten.columns: # dentro do DataFrame(train), ele irá olhar coluna por coluna quais tipos elas são
        dtype = str(rotten[col].dtype)
        types.append(dtype)
    total_percentual['Types'] = types
    return(np.transpose(total_percentual))
dados_ausentes(rotten)
```
